```python
import pandas as pd

# Extract 'SignupYear' from 'SignupDate'
data['SignupDate'] = pd.to_datetime(data['SignupDate'])
data['SignupYear'] = data['SignupDate'].dt.year

# Group by 'CustomerID' to calculate 'TotalSpend' and 'NumPurchases'
customer_profile = data.groupby('CustomerID').agg(
    SignupYear=('SignupYear', 'first'),  # Take the first (all customers have the same year)
    TotalSpend=('TotalValue', 'sum'),
    NumPurchases=('TransactionID', 'count')
).reset_index()

# Check the new customer profile
print(customer_profile.head())

# Proceed with clustering
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Features for clustering
data_features = customer_profile[['SignupYear', 'TotalSpend', 'NumPurchases']]

# Scaling the features
scaler = StandardScaler()
data_features_scaled = scaler.fit_transform(data_features)

# Apply KMeans clustering (you can change the number of clusters between 2 and 10)
kmeans = KMeans(n_clusters=4, random_state=42)
customer_profile['Cluster'] = kmeans.fit_predict(data_features_scaled)

# Visualize the clusters (using 2D plot for example)
import matplotlib.pyplot as plt

plt.scatter(customer_profile['TotalSpend'], customer_profile['NumPurchases'], c=customer_profile['Cluster'], cmap='viridis')
plt.xlabel('Total Spend')
plt.ylabel('Number of Purchases')
plt.title('Customer Segmentation')
plt.colorbar(label='Cluster')
plt.show()

# Calculate DB Index to evaluate clustering quality
from sklearn.metrics import davies_bouldin_score
db_index = davies_bouldin_score(data_features_scaled, customer_profile['Cluster'])
print(f"DB Index: {db_index}")
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

clusters_range = range(2, 11)  # Range of k (2 to 10 clusters)
inertia = []
db_indices = []

for k in clusters_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data_scaled)

    # Calculate inertia and Davies-Bouldin Index for each k
    inertia.append(kmeans.inertia_)
    db_indices.append(davies_bouldin_score(data_scaled, kmeans.labels_))

# Plot Inertia (Elbow Curve)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(clusters_range, inertia, marker='o')
plt.title("Elbow Method")
plt.xlabel("Number of clusters")
plt.ylabel("Inertia")

# Plot DB Index for evaluation
plt.subplot(1, 2, 2)
plt.plot(clusters_range, db_indices, marker='o')
plt.title("Davies-Bouldin Index")
plt.xlabel("Number of clusters")
plt.ylabel("DB Index")
plt.tight_layout()
plt.show()
```

```python
# Inertia (within-cluster sum of squares)
print(f"Inertia: {kmeans_final.inertia_}")

# Silhouette Score
from sklearn.metrics import silhouette_score
silhouette_avg = silhouette_score(data_scaled, data['Cluster'])
print(f"Silhouette Score: {silhouette_avg}")

# Cluster Centers (Centroids)
centroids = kmeans_final.cluster_centers_
print(f"Cluster Centroids:\n {centroids}")
```
